```python
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt

RESULT_DIR = '../bin'

result_df = pd.read_csv(f'{RESULT_DIR}/gens_sample_time.csv')

fig, ax = plt.subplots(3, 3, figsize=(30, 16))

model_names = list(set(result_df.model_label.unique()))
colors = ['b', 'g', 'r', 'c']

for i, n in enumerate(result_df.num_samples.unique()):
    row = i // 3
    col = i % 3

    sub_df = result_df[result_df.num_samples == n]
    dfs = [sub_df[sub_df.model_label == l] for l in model_names]

    for d, l, c in zip(dfs, model_names, colors):
        ax[row, col].plot(d.num_threads, d.generate_time, color=c, marker='.', label=l)
        ax[row, col].set_title(f'num_samples={n}')
        ax[row, col].set_xlabel('num_threads')
        ax[row, col].set_ylabel('time (s)')
        ax[row, col].legend()
```
